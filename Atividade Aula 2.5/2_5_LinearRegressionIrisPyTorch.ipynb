# Regressão Linear com PyTorch

## Objetivo

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se
- o cálculo do gradiente automático do PyTorch, porém
- a função de perda e
- atualização dos parâmetros é feita de forma explícita.
- A rede é criada com uma camada nn.Linear()

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinear_Otimizacao.png)

## Importação dos pacotes

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn.datasets import load_iris

import torch
from torch import nn, optim
from torch.autograd import Variable

torch.manual_seed(1234)
```

## Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias
```

```python
print('x_train_bias.shape:',x_train_bias.shape, x_train_bias.dtype)
```

```python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
print(y_train)
```

## Model

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinearNeuronio.png)

```python
#Inicialização usando toch.nn.Linear

model = torch.nn.Linear(in_features=2, out_features=1, bias=False)
```

### Verificando a inicialização dos parâmetros

```python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight
```

```python
#Inicialização usando toch.nn.Linear

model2 = torch.nn.Linear(in_features=1, out_features=1, bias=True)
```

### Testando o predict da rede

```python
xx = torch.ones((5,2))
xx
```

```python
model(Variable(xx))
```

```python
model(x_train_bias)
```

## Treinamento

```python
num_epochs = 100
learning_rate = 0.5
for epoch in range(num_epochs):
    #inputs = Variable(x_train_bias)
    #target = Variable(y_train)
    inputs = x_train_bias
    target = y_train

    # forward - predict
    y_pred = model(inputs)

    #loss cálculo da função de perda
    loss = torch.mean((y_pred - target)**2)

    # backward propagation - cálculo gradient
    loss.backward()

    # gradiente descendente
    model.weight.data = model.weight.data - learning_rate * model.weight.grad.data

    # zeragem parâmetros próximo cálculo do gradiente
    model.weight.grad.data.zero_()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
```

## Avaliação

```python
y_pred = model(Variable(x_train_bias))
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
```

### Parâmetros ótimos, solução analítica

```python
x_bias = x_train_bias
y = y_train
x_bias_t = torch.t(x_bias)
w_opt = (torch.inverse(x_bias_t.mm(x_bias)).mm(x_bias_t)).mm(y)
print(torch.t(w_opt))
```

# Exercícios

- 1) Por que no final da varredura do laço de treinamento, a perda é impressa utilizando
   `loss.data`? Por que precisa do `data`?

   O loss é o valor da perda mais um registro completo, o histórico de todas as operações matemáticas (y_pred, mean, square, etc.)\
   O .data é tipo "Pegue apenas o número e ignore o histórico de cálculo", dizendo que está apenas lendo o valor.
   

- 2) Calcule e imprima o valor da perda (MSE) fazendo o predict da rede com os pesos já

- 3) Crie duas listas `w0_list` e `w1_list` para guardar a evolução dos valores dos
   parâmetros durante o laço de treinamento.
   Modifique o exemplo do gráfico tipo **scatter plot** para visualizar a variação

```python
# ----- Questão 2 --------
# Realizar o predict com os dados de treino e o modelo já treinado
y_pred_trained = model(x_train_bias) 

# 2. Calcular a perda (MSE)
loss_trained = torch.mean((y_pred_trained - y_train)**2)

# Imprimir o valor da perda
print(f"Perda (MSE) com pesos treinados: {loss_trained.data.item():.16f}")
```

```python
# ----- Questão 3 --------
# O vetor ótimo w_opt é :
# [[0.2613, 0.6865]]
# Onde [0.2613 é o w0 (bias) e 0.6865 é o w1]
w_opt_numpy = np.array([[0.2613, 0.6865]]) 
w0_opt = w_opt_numpy[0, 0]
w1_opt = w_opt_numpy[0, 1]

# Recriar e inicializar o modelo
model_plot = torch.nn.Linear(in_features=2, out_features=1, bias=False)
model_plot.weight.data = torch.zeros(1, 2)
torch.nn.init.uniform_(model_plot.weight.data, -0.1, 0.1) 

num_epochs = 100
learning_rate = 0.5
x_train_bias = x_train_bias 
y_train = y_train   

#Listas para guardar a evolução dos parâmetros
w0_list = []
w1_list = []

for epoch in range(num_epochs):
    inputs = x_train_bias
    target = y_train

    # forward - predict
    y_pred = model_plot(inputs)

    #loss cálculo da função de perda
    loss = torch.mean((y_pred - target)**2)

    # backward propagation - cálculo gradient
    loss.backward()

    # Guarda os pesos antes de atualizar
    w0_list.append(model_plot.weight.data[0, 0].item())
    w1_list.append(model_plot.weight.data[0, 1].item())

    # gradiente descendente
    model_plot.weight.data = model_plot.weight.data - learning_rate * model_plot.weight.grad.data

    # zeragem parâmetros
    model_plot.weight.grad.data.zero_()
    
# --- Plotagem da Evolução dos Parâmetros ---
plt.figure(figsize=(7, 5))

# 1. Trajetória
w0_old, w1_old = w0_list[0], w1_list[0]
for w0, w1 in zip(w0_list[1:], w1_list[1:]):
    plt.arrow(w0_old, w1_old, w0 - w0_old, w1 - w1_old,
              head_length=0.005, head_width=0.002, shape='full',
              length_includes_head=True, color='blue', alpha=0.5)
    w0_old, w1_old = w0, w1

# 2. Scatter dos pontos
plt.scatter(w0_list, w1_list, s=10, c=range(num_epochs), cmap='viridis', label='Trajetória do GD')

# 3. Ponto ótimo analítico
plt.scatter(w0_opt, w1_opt, s=100, c='red', marker='*', label='Ótimo Analítico')

plt.title('Trajetória dos Parâmetros (w0 vs w1)')
plt.xlabel('w0 (Bias)')
plt.ylabel('w1 (Inclinação)')
plt.legend()
plt.grid(True)
plt.show()
```
